```python
#Returns the dictionary to the calling statement 
def fit(x,y):
    counts={}
    class_values=set(y)       #Contains values of 20 classes
    for current_class in class_values:
        total=0
        counts[current_class]={}
        rows_with_current_class=(y==current_class)
        x_current_class=x[rows_with_current_class]
        y_current_class=y[rows_with_current_class]
        for i in range(x_current_class.shape[0]):
            for j in range(x_current_class.shape[1]):
                total=total+x_current_class[i][j]
        counts[current_class]["Total count"]=total
        for i in range(x_current_class.shape[1]):  #x_current_class.shape[1] will give 3000 
            z=0
            for j in range(x_current_class.shape[0]):     #x_current_class.shape[0] will give the no of rows to add the frequencies
                z=z+x_current_class[j][i]
            counts[current_class][i]=z
    return counts
```

```python
def predict(x_test,counts):          #x_test received has a shape of (500,3000)
    output=[]
    for x in x_test:
        output.append(predict_single(x,counts))
    return output
```

```python
def predict_single(x,counts):        # x is a tuple of x_test containing fequency of 3000 words in a single row
    max_pr=float("-inf")
    classes=counts.keys()
    for c in classes:         #Loop to iterate over all the 20 classes and return the class having maximum probability
        p=findSingleClassProb(x,counts,c)
        if p > max_pr:
            max_pr=p
            max_class=c
    return max_class
```

```python
def findSingleClassProb(x,counts,c):
    ans=0   
    for i in range(len(x)):       #loop to iterate over all the 3000 words of a tuple
        #if(x[i]>0):
        prob=math.log(counts[c][i]+1)-math.log(counts[c]["Total count"]+3000)
        log_prob=prob*x[i]
        ans=ans+(log_prob)
    return ans
```
